```python
def printElements(s,info):
    print(info)
    if len(s)==0:
        print("Himpunan kosong")
    for i in s:
        print(i)

s = set([11,22,33,44,55])
printElements(s,"Sebelum:")

s.add(66)
s.add(77)
s.add(88)
s.add(99)
s.add(88)
print(s)
printElements(s,"Setelah pemanggilan add():")

# update
s.update([2025])
printElements(s,"Setelah pemanggilan update():")

# menghapus elemen
s.remove(55)
printElements(s,"Setelah pemangilan remove():")

# menghapus semua elemen
s.clear()
printElements(s,"Setelah pemanggilan clear():")

x = set([1,2,3])
y = set([3,4])

# union (Menggambungkan)
union = x | y
print("hasil union dari 2 seet:", union)

# intersection (data yang sama di kedua set)
inter_section = x & y
print("hasil intersection dari 2 set:", inter_section)

# operation difference (ada di set awal tapi tidak ada di set yang lain)
difference = x - y
print("hasil difference dari 2 set:", difference)

# symentric diffrence (data yang beda di kedua set, namun yang sama akan dibuang)
symetric_difference = x ^ y
print("hasil symetric difference dari 2 set:", symetric_difference)
```
