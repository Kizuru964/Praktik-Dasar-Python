```python
# Tipe string
# Menampilkan karakter pertama, 
#diawali dari 0
print(' ')
s = 'Takumi'
print(s[0])
print(s[1])
print(s[2])
print(s[3])
print(s[4])
print(s[5])
```

```python
# Menampilkan karakter dari kanan ke kiri, 
# dimulai dari s[-1]
print(' ')
print(s[-1])
print(s[-2])
print(s[-3])
print(s[-4])
print(s[-5])
print(s[-6])
print(' ')
# Jika disusun dari s[-6] 
print(' ')
print(s[-6])
print(s[-5])
print(s[-4])
print(s[-3])
print(s[-2])
print(s[-1])
```

```python
# Menampilkan dua karakter di awal
print(' ')
print(s[:2])
```

```python
# Menampilkan tiga karakter terakhir
print(s[3:]) 
```

```python
# Membandingkan string
s1 = 'Halo' #variabel string 1
s2 = 'Takumi' #variabel string 2
print(s1 > s2)
print(s1 >= s2)
print(s1 < s2)
print(s1 <= s2)
print(s1 != s2)
```

```python
#Type integer
a = 1
b = -17
c = 100

print(type(a))
print(type(b))
print(type(c))
```

```python
#Type float
x = 3.14
y = -0.175
z = 2.05

print(type(x))
print(type(y))
print(type(z))
```

```python
#Type complex
s1 = 3 + 2j
s2 = -2.5j

print(type(s1))
print(type(s2))
print(s1.real) #menampilkan bilangan real
print(s1.imag) #menampilkan bilangan imajer
```

```python
#Konversi Tipe Data Numerik

#Konversi ke integer
x = int(3.14)
print(x)

#Konversi ke float
b = float(-17)
print(b)
#Konversi ke complex
y = complex(-0.175)
print(y)
```

```python
#Operasi Aritmatika
a = 15
b = 6
print(a + b) #penjumlahan
print(a - b) #pengurangan
print(a * b) #perkalian
print(a / b) #[pembagian]
print(a // b) #untuk pembagian bilangan bulat
print(a % b) #modulus
print(a ** b) #eskponen
```

```python
#Modul math
import math
a = 4
b = 2
c = 0
print(math.sqrt(a)) #akar variabel a = 4
print(math.pi*(b)) #perkalian variabel b = 2 dengan pi = 3.14
print(math.cos(c)) #cosinus variabel c = 0
```

```python
#Modul cmath
import cmath
z = 1 + 4j
print(cmath.sqrt(z)) #akar variabel z
print(cmath. sin(z)) #sinus variabel z
```
