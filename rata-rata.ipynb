```python
# Definisi Data Mahasiswa
nama = ["Bahar", "Ruddi", "Ghalib", "Kizuru"]
prakA = [70, 60, 70, 65]
prakB = [80, 90, 85, 80]

# Hitung Rata-Rata Keseluruhan

# Menggabungkan semua nilai prakA dan prakB menjadi satu list
semua_nilai = prakA + prakB

# Hitung Total dan Jumlah Angka
total_nilai = sum(semua_nilai)
jumlah_data = len(semua_nilai)

# Hitung Rata-Rata Keseluruhan
rata_rata_keseluruhan = total_nilai / jumlah_data

print("=========================================")
print(f"Rata-Rata Keseluruhan Nilai (Batas Lulus): **{rata_rata_keseluruhan:.2f}**")
print("=========================================\n")


# --- 3. Iterasi dan Evaluasi Status Kelulusan ---

# Menggunakan zip untuk mengulang melalui nama, prakA, dan prakB secara bersamaan
for n, nilai_a, nilai_b in zip(nama, prakA, prakB):
    
    status = ""
    
    # Kriteria baru: Mahasiswa LULUS jika nilai_a ATAU nilai_b lebih dari rata-rata
    if nilai_a > rata_rata_keseluruhan and nilai_b > rata_rata_keseluruhan:
        status = "LULUS"
    else:
        status = "NILAI MASIH ADA YANG KURANG"
        
    # Tampilkan hasil per mahasiswa
    print(f"Nama Mahasiswa: {n}")
    print(f"   Nilai Prak A: {nilai_a}")
    print(f"   Nilai Prak B: {nilai_b}")
    print(f"   Status: **{status}**")
    print("-" * 20)
```
