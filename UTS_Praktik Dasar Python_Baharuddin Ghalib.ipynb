```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import StateSpace, ss2tf, step

A = np.array([[0, 1],
              [-1, -3]])
B = np.array([[0, 0],
              [2, 4]])
C = np.array([[1, 0]])
D = np.array([[0, 0]])

# Transfer function
num1, den = ss2tf(A, B, C, D, input=0)
num2, _   = ss2tf(A, B, C, D, input=1)

# Sistem untuk step response
sys1 = StateSpace(A, B[:,0:1], C, [[0]])
sys2 = StateSpace(A, B[:,1:2], C, [[0]])

t1, y1 = step(sys1)
t2, y2 = step(sys2)

plt.plot(t1, y1, label='H1(s)')
plt.plot(t2, y2, label='H2(s)')
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Response')
plt.title('Perbandingan Step Response')
plt.grid(True)
plt.show()
```
