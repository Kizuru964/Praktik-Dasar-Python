```python
import random

lux_value = random.uniform(0, 2000) # Menghasilkan nilai lux acak antara 0 hingga 2000 lux

keterangan = '' # Inisialisasi variabel keterangan

# Meenentuan Keterangan Lampu
if lux_value < 50:
    keterangan = 'Gelap.'
elif lux_value < 150:
    keterangan = 'Cahaya redup.'
elif lux_value < 300:
    keterangan = 'Pencahayaan sedang.'
elif lux_value < 500:
    keterangan = 'Terang.'
elif lux_value < 1000:
    keterangan = 'Sangat terang.'
else: # Jika lux_value >= 1000
    keterangan = "Sangat cerah."

# Menampilkan Hasil
print("--- Hasil Pengukuran Simulasi ---")
print(f"Nilai Intensitas Cahaya (Lux): {lux_value:.2f} lux")
print(f"Keterangan Lampu: **{keterangan}**")
```
